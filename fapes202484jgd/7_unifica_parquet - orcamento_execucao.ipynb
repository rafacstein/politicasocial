```python
import pandas as pd

# List of input Parquet files
parquet_files = [
    "C:/Users/RafaelStein/Documents/file_1.parquet",
    "C:/Users/RafaelStein/Documents/file_2.parquet",
    "C:/Users/RafaelStein/Documents/file_3.parquet",
    "C:/Users/RafaelStein/Documents/file_4.parquet",
    "C:/Users/RafaelStein/Documents/file_5.parquet",
    "C:/Users/RafaelStein/Documents/file_6.parquet",
    "C:/Users/RafaelStein/Documents/file_7.parquet",
    "C:/Users/RafaelStein/Documents/file_8.parquet",
    "C:/Users/RafaelStein/Documents/file_9.parquet",
    "C:/Users/RafaelStein/Documents/file_10.parquet",
    "C:/Users/RafaelStein/Documents/file_11.parquet",
    "C:/Users/RafaelStein/Documents/file_12.parquet",
    "C:/Users/RafaelStein/Documents/file_13.parquet",
    "C:/Users/RafaelStein/Documents/file_14.parquet",
    "C:/Users/RafaelStein/Documents/file_15.parquet"
]

# Output location
output_file = "C:/Users/RafaelStein/Downloads/orcamento_execucao_2009_2024.parquet"

# This list will hold the file paths for intermediate results
intermediate_files = []

# Process each Parquet file one by one
for i, file in enumerate(parquet_files):
    try:
        # Read Parquet file
        df = pd.read_parquet(file)

        # Convert all columns to strings (efficiently one by one)
        for col in df.columns:
            df[col] = df[col].astype(str)

        # Generate intermediate output file for each Parquet file processed
        intermediate_file = f"C:/Users/RafaelStein/Downloads/intermediate_{i}.parquet"
        intermediate_files.append(intermediate_file)
        df.to_parquet(intermediate_file, engine='pyarrow')  # You can try 'fastparquet' here too

        print(f"Processed and saved: {intermediate_file}")

    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Now, concatenate all intermediate files
try:
    # Read all intermediate Parquet files into a single DataFrame
    intermediate_dfs = [pd.read_parquet(file) for file in intermediate_files]
    combined_df = pd.concat(intermediate_dfs, ignore_index=True)

    # Save the combined DataFrame as a new Parquet file
    combined_df.to_parquet(output_file, engine='pyarrow')
    print(f"Combined Parquet file saved at: {output_file}")

except Exception as e:
    print(f"Error combining files: {e}")
```
